```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['Simhei']  #显示中文
plt.rcParams['axes.unicode_minus']=False    #显示负号
%config InlineBackend.figure_format = 'retina'
plt.style.use('ggplot')

import scipy.interpolate as spip
```

```python
x = np.linspace(0, 10, 11)
xnew = np.linspace(0, 10, 10001)
y = np.cos(-x**2 / 9.0)
yorgin = np.cos(-xnew**2 / 9.0)
fig,ax = plt.subplots()
ax.scatter(x,y,label = '采样数据')
ax.plot(xnew,yorgin , label = '原始数据')
ax.legend()
None
```

```python
ynew = np.interp(xnew, x, y)
ax.plot(xnew,ynew,label = '线性插值')
ax.legend()
fig
```

```python
spl = spip.CubicSpline(x,y)
ax.plot(xnew,spl(xnew),label='三次样条')
ax.legend()
display(fig)
```

```python
ax_CubicSpline = plt.subplot()
ax_CubicSpline.scatter(x,y,label = '采样数据')
ax_CubicSpline.plot(xnew,yorgin , label = '原始数据')

ax_CubicSpline.plot(xnew,spl(xnew,nu=1),label='三次样条插值的一阶导数')
ax_CubicSpline.plot(xnew,spl(xnew,nu=2),label='三次样条插值的二阶导数')
ax_CubicSpline.plot(xnew,spl(xnew,nu=3),label='三次样条插值的三阶导数')
ax_CubicSpline.plot(xnew,spl(xnew,nu=4),label='三次样条插值的四阶导数')
ax_CubicSpline.legend()
None
```

```python
ax.plot(xnew,spip.make_interp_spline(x,y,2)(xnew),label='B样条(k=2)')
ax.plot(xnew,spip.make_interp_spline(x,y,3)(xnew),label='B样条(k=3)（等于三次样条）')
ax.plot(xnew,spip.make_interp_spline(x,y,5)(xnew),label='B样条(k=5)')
ax.plot(xnew,spip.make_interp_spline(x,y,7)(xnew),label='B样条(k=7)')
ax.legend()
display(fig)
```
