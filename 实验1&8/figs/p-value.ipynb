```python
from tikz import *
import numpy as np
from scipy.stats import chi2
```

```python
chi2(2).pdf(0)*1.2
```

```python
eps = 0
x=np.linspace(eps,5,100)

hight = chi2(2).pdf(0)*1.2

pic = Picture(x='2cm',y='6cm')

# axis code
scope = pic.scope()

# horizontal axis and label
scope.draw('(-10pt, 0)', lineto((5.1, 0)), node('$x$', right=True), coordinate(name='x axis'),
    opt='-latex')

# vertical axis and label
scope.draw('(0,-10pt)', lineto((0, chi2(2).pdf(0)*1.2)), node(r'$f_{\chi^2}(x)$', above=True), coordinate(name='y axis'),
    opt='-latex')


pic.draw(line(list(zip(x,chi2(2).pdf(x)))))

y=np.linspace(2,5,100)
pic.fill((2,0),
         lineto(list(zip(y,chi2(2).pdf(y)))),
         lineto((5,0)),
         '--cycle',fill='red',fill_opacity=0.3)
pic.node(r'$\alpha$',at=(2,chi2(2).pdf(2)/2),right=True)
pic.draw((2,0),node(r'$\chi^2_{\alpha}$',below=True)
         ,lineto((2,chi2(2).pdf(2))),
         dashed=True)

y=np.linspace(3,5,100)
pic.fill((3,0),
         lineto(list(zip(y,chi2(2).pdf(y)))),
         lineto((5,0)),
         '--cycle',fill='gray',fill_opacity=0.7)
pic.node(r'$p$-value',at=(3,chi2(2).pdf(3)/2),right=True)
pic.draw((3,0),node(r'$\chi^2$',below=True),
         lineto((3,chi2(2).pdf(3))),
         dashed=True)

# pic.demo(300)
pic.write_image('p-value in chi2 test.pdf')
print(pic.code())
pic
```
