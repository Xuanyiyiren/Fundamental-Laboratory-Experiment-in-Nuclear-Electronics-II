```python
from tikz import *
```

```python
ep7 = Picture()
origin = (0,0)
distance = 7

ep7.usepackage('ctex','UTF8') # 使用中文

ep7.draw((origin[0]-1,origin[1]+distance/2-0.2),rectangle((origin[0]+1,origin[1]+distance/2+0.2)))
ep7.node('跟随器','GSQ',(origin[0],origin[1]+distance/2))
ep7.node('跟随器','GSQ',(origin[0],origin[1]-distance/2))

print(ep7.document_code())
ep7
```
